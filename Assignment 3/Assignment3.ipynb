```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
row1 = df_encoded.iloc[0].values.reshape(1,-1)
row2 = df_encoded.iloc[1].values.reshape(1,-1)

# Cosine Similarity
cos_sim = cosine_similarity(row1, row2)[0][0]
```

```python
# Jaccard Similarity (for binary attributes only)
def jaccard(a,b):
    intersection = np.logical_and(a,b).sum()
    union = np.logical_or(a,b).sum()
    return intersection/union

jacc = jaccard(row1[0], row2[0])
```

```python
def smc(a,b):
    return (a==b).sum()/len(a)

smc_val = smc(row1[0], row2[0])
```

```python
print("Cosine:", cos_sim, "Jaccard:", jacc, "SMC:", smc_val)
```
